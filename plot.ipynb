```python
import matplotlib.pyplot as plt
import numpy as np
def plot_scores(
    models,
    precisions, 
    accuracies, 
    fmeasures, 
    recalls
):
    fig, axes = plt.subplots(2, 2, figsize=(15, 15))

    axes = axes.reshape(-1)
    fig.subplots_adjust(hspace=0.5, wspace=0.5)
    
    for ax, score, score_label in zip(axes, [precisions, accuracies, fmeasures, recalls], ["weighted precision avg", "accuracy", "weighted f1-score avg", "weighted recall avg"]):
        ax.set_xlabel("Models")
        ax.set_ylabel(score_label)


        ax.set_ylim(0, 1)

        X_axis = np.arange(len(models))

        # set xticks
        ax.set_xticks(X_axis)
        ax.grid()
        ax.set_xticklabels(models, rotation=90)

        ax.bar(
            X_axis -0.2, score[0],  color="#0288D1dd", label="TopHits", width=0.4
        )
        ax.bar(
            X_axis + 0.2, score[1],  color="#388E3C", label="Fetched", width=0.4
        )
        

        ax.legend(loc="best")

    return plt
```
